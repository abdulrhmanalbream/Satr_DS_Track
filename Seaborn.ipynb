```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))

sns.histplot(x = tips.tip, hue = tips.day, element='step')
plt.title('Histogram of Tips by Day')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.displot(x = tips.day, hue = tips.sex, multiple='dodge', shrink = 0.8 ,
      palette= 'Set1', hue_order = ['Female', 'Male'])
plt.title('Displot of Days by gender')
plt.show()
```

```python
# ecdf plot --------> tip with time
sns.set_style('darkgrid')
sns.set(font_scale=1.25)

plt.figure(figsize=(10, 6))
sns.ecdfplot(x = tips.tip, hue=tips.time, palette='summer', linewidth=3)
plt.title('ECDF of Tips by Time')
plt.show()
```
