```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([1, 2, 3, 5, 8, 13, 20])

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
plt.plot(x,  marker='o', linestyle='--', linewidth = '3', markersize = 10)
plt.plot(y,  marker='o', linestyle='--', linewidth = '3', markersize = 10)

plt.legend(['Normal','Fast']);
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#super title = my data visulaization assignment

#title = First Chart
x_plot1 = np.array([1, 2, 3, 4, 5, 6, 7])
y_plot1 = np.array([1, 1, 2, 3, 5, 8, 13])

#title = Second Chart
x_plot2 = np.array([0, 1, 2, 3, 4, 5, 6])
y_plot2 = np.array([2, 4, 6, 8, 10, 12, 14])

#title = Third Chart
x_plot3 = np.array([0, 1, 3, 4])
y_plot3 = np.array([4, 6, 3, 4])
```

```python
plt.subplots(1, 3, figsize=(15, 5))

#First Chart
plt.subplot(1, 3, 1)
plt.plot(x_plot1,y_plot1)
plt.title('First Chart')

#Second Chart
plt.subplot(1, 3, 2)
plt.plot(x_plot2,y_plot2)
plt.title('Second Chart')

#Third Chart
plt.subplot(1, 3, 3)
plt.plot(x_plot3,y_plot3)
plt.title('Third Chart')

plt.suptitle("my data visulaization assignment ")

plt.show()
```
